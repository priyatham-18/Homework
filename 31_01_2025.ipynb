```python
# prompt: how to upload a csv file

from google.colab import files
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
#Then you can use pandas to read the csv file
import pandas as pd
df = pd.read_csv(fn)
print(df.head())
```

```python
import seaborn as sns
sns.distplot(df['daily'])
```

```python
#fitting a linear regression model
import statsmodels.formula.api as smf
model = smf.ols('sunday~daily',data=df).fit()
sns.regplot(x='daily',y='sunday',data=df)
```

```python
# prompt: implement histogram

import matplotlib.pyplot as plt

# Assuming 'df' and 'daily' column exist from previous code
plt.hist(df['daily'], bins=10)  # Adjust the number of bins as needed
plt.xlabel('Daily Values')
plt.ylabel('Frequency')
plt.title('Histogram of Daily Values')
plt.show()
```
