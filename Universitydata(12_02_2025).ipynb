```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.cluster import DBSCAN
```

```python
file_path = '/content/drive/MyDrive/Universities.csv'
df =pd.read_csv(file_path)
```

```python
df
```

```python
df.drop(['Univ', 'SAT' ],axis=1,inplace=True)
```

```python
array=df.values
```

```python
array
```

```python
stscalear=StandardScaler().fit(array)
X=stscalear.transform(array)
```

```python
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([df,cl],axis=1)
```
